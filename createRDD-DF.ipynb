```
#empty RDD
emptyrdd = spark.sparkContext.emptyRDD()
print(emptyrdd.collect())
```

```
#empty rdd using parallelize
rdd2 = spark.sparkContext.parallelize([])
print(rdd2.collect())
```

```
#empty dataframe
from pyspark.sql.types import *
schema = StructType([StructField("Name",StringType(),True),StructField("Address",StringType(),True)])
df = spark.createDataFrame(rdd2,schema)
df.printSchema()
display(df)
```

```
#empty df with rdd
df = rdd2.toDF(schema)
```
