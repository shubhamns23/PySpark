```
%python
simpleData = [("James", "Sales", 3000),
    ("Michael", "Sales", 4600),
    ("Robert", "Sales", 4100),
    ("Maria", "Finance", 3000),
    ("James", "Sales", 3000),
    ("Scott", "Finance", 3300),
    ("Jen", "Finance", 3900),
    ("Jeff", "Marketing", 3000),
    ("Kumar", "Marketing", 2000),
    ("Saif", "Sales", 4100)
  ]
schema = ["employee_name", "department", "salary"]
df = spark.createDataFrame(data=simpleData, schema = schema)
df.printSchema()
display(df)
```

```
from pyspark.sql.functions import *
from pyspark.sql.window import *
winspec = Window.partitionBy("department").orderBy("salary")
df.withColumn("row_number",row_number().over(winspec)).show()
```

```
df.withColumn("rank",rank().over(winspec)).show()
```

```
df.withColumn("dense_rank",dense_rank().over(winspec)) \
    .show()
```

```
df.withColumn("percent_rank",percent_rank().over(winspec)) \
    .show()
```

```
df.withColumn("ntile",ntile(3).over(winspec)) \
    .show()
```
