```
data  = [("James","Smith","USA","CA"),("Michael","Rose","USA","NY"),("Robert","Williams","USA","CA"),("Maria","Jones","USA","FL")]
cols = ["firstname","lastname","country","state"]
df = spark.createDataFrame(data,cols)
df.printSchema()
display(df)
```

```
df.select(df.state).show()
df.select(df["state"]).show()
from pyspark.sql.functions import *
df.select(col("state")).show()
df.select("state").show()
```

```
state1 = df.rdd.map(lambda x: x[3]).collect()
print(state1)
```

```
state2 = df.rdd.map(lambda x: x.state).collect()
print(state2)
```

```
state3 = df.select("state").rdd.flatMap(lambda x: x).collect()
print(state3)
```
