```
#create RDD
data = [("Finance",10),("Marketing",20),("Sales",30),("IT",40)]
rdd = spark.sparkContext.parallelize(data)
```

```
#using toDF
df = rdd.toDF()
df.printSchema()
display(df)
```

```
#using columns toDF
cols = ["Team","Count"]
df = rdd.toDF(cols)
df.printSchema()
display(df)
```

```
#using create function
df = spark.createDataFrame(rdd,cols)
df.printSchema()
```

```
#using StructType
from pyspark.sql.types import *
schema = StructType().add("Team",StringType(),True).add("Count",IntegerType(),True)
df = spark.createDataFrame(rdd,schema)
df.printSchema()
df.show()
```

```
#using csv file
df = spark.read.format('csv').option('header','true').option('inferschema','true').load('/FileStore/movies.csv')
df.printSchema()
display(df)
```
