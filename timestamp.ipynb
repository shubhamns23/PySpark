```
data = [("1","2019-06-24 12:01:19.000")]
cols = ["Id","input"]
df = spark.createDataFrame(data,cols)
df.printSchema()
display(df) 
```

```
df.select("Id",to_timestamp("input").alias("Time"),current_timestamp(),to_date("input")).show(truncate=False)
```

```
 df=spark.createDataFrame(
        data = [ ("1","2019-06-24 12:01:19.000")],
        schema=["id","input_timestamp"])
df.printSchema()
```

```
df.withColumn("date_type",to_date("input_timestamp")) \
  .show(truncate=False)
```

```
df.select(to_date(lit('06-24-2019 12:01:19.000'),'MM-dd-yyyy HH:mm:ss.SSSS')) \
  .show()
```

```
df.withColumn("ts",to_timestamp("input_timestamp")).withColumn("date",to_date("ts")).show()
```

```
df.withColumn("ts",df.input_timestamp.cast('timestamp')).withColumn("date",df.input_timestamp.cast('date')).show()
```
