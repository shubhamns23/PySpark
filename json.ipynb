```
jsonString="""{"Zipcode":704,"ZipCodeType":"STANDARD","City":"PARC PARQUE","State":"PR"}"""
df=spark.createDataFrame([(1, jsonString)],["id","value"])
df.show(truncate=False)
```

```
from pyspark.sql.types import *
df2 = df.withColumn("value",from_json("value",MapType(StringType(),StringType())))
display(df2)
df2.show()
```

```
from pyspark.sql.types import *
schema = StructType()\
    .add("Zipcode",IntegerType())\
    .add("ZipCodeType",IntegerType())\
    .add("City",StringType())\
    .add("State",StringType())
df2 = df.withColumn("value",from_json("value",schema))
display(df2)
```

```
df3 = df2.select("id","value.*")
display(df3)
```

```
df2.withColumn("Value",to_json("value")).show()
```

```
df.select("id",json_tuple("value","Zipcode","ZipCodeType","City")).toDF("id","Zipcode","ZipCodeType","City").show()
```

```
df.select("id",get_json_object("value","$.ZipCodeType").alias("ZipCode Type")).show()
```
