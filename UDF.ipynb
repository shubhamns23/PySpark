```
columns = ["Seqno","Name"]
data = [("1", "john jones"),
    ("2", "tracey smith"),
    ("3", "amy sanders")]

df = spark.createDataFrame(data=data,schema=columns)

df.show(truncate=False)
```

```
def tuper(string):
    return string.upper()
tuper('Shubham')
```

```
from pyspark.sql.functions import col, udf
upperf = udf(tuper)
```

```
df.select(df['Seqno'],upperf(df['Name'])).show()
```

```
def convert(str):
    res = ""
    arr = str.split(" ")
    for name in arr:
        res = res + name[0:1].upper()+""+name[1:len(name)]+" "
    return res
convert('shubham salve')    
```

```
convertf = udf(convert)
```

```
df.select('Seqno',convertf('Name').alias('Name')).show()
```
